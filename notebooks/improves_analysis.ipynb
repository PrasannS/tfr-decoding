```python
improvedset = pd.read_json("output/std_dataset.jsonl", orient='records', lines=True)
improvedset['maximp'] = improvedset['max_scos_best'] - improvedset['max_scos_first']
improvedset['meanimp'] = improvedset['mean_scos_best'] - improvedset['mean_scos_first']
```

```python
def boundary_confusion_matrix(df, col1, col2, threshold1, threshold2):
    # Initialize the confusion matrix
    matrix = np.zeros((2, 2))

    # Calculate the confusion matrix
    for _, row in df.iterrows():
        if row[col1] < threshold1:
            if row[col2] < threshold2:
                matrix[0][0] += 1
            else:
                matrix[0][1] += 1
        else:
            if row[col2] < threshold2:
                matrix[1][0] += 1
            else:
                matrix[1][1] += 1

    return matrix
```

```python
boundary_confusion_matrix(improvedset, "meanimp", "mean_scos_first", 0.05, 0.8)
```
